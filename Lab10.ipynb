```python
import pandas as pd
import statistics as st
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
from scipy.stats import pearsonr, spearmanr, kendalltau, norm
import seaborn as sns
```

```python
df = pd.read_csv('E:\\Магистратура\\3\\Машинное обучение\\healthcare-dataset-stroke-data.csv', sep=',')
print(df.head(20))
# print(df.info())
# print(df.shape)
# print(df.stroke.unique())
```

```python
# Центральные метрики

glucose = df['avg_glucose_level']
age = df['age']
fig, ax = plt.subplots()
ax.scatter(glucose, age, c='red', alpha=0.3, edgecolors='none')
ax.grid(True)
plt.xlabel('avg_glucose_level')
plt.ylabel('age')
plt.show()

print("\nМатематическое ожидание: ")
print(st.mean(df['avg_glucose_level']))

print("\nСредневзвешенное значение: ")
print(np.average(df['avg_glucose_level']))

print("\nМедиана: ", st.median(df['avg_glucose_level']))
print("Медиана меньшее среднее значение: ", st.median_low(df['avg_glucose_level']))
print("Медиана большее среднее значение: ", st.median_high(df['avg_glucose_level']))

print("\nМода: ")
print('Значение моды: ', st.mode(df['avg_glucose_level']))

print("\nМинимум: ", np.min(df['avg_glucose_level']))
print("Максимум: ", np.max(df['avg_glucose_level']))
```

```python
print("\nМатематическое ожидание: ")
print(st.mean(df['id']))

print("\nСредневзвешенное значение: ")
print(np.average(df['id']))

print("\nМедиана: ", st.median(df['id']))
print("Медиана меньшее среднее значение: ", st.median_low(df['id']))
print("Медиана большее среднее значение: ", st.median_high(df['id']))

print("\nМода: ")
print('Значение моды: ', st.mode(df['id']))

print("\nМинимум: ", np.min(df['id']))
print("Максимум: ", np.max(df['id']))
```

```python
print("\nМатематическое ожидание: ")
print(st.mean(df['age']))

print("\nСредневзвешенное значение: ")
print(np.average(df['age']))

print("\nМедиана: ", st.median(df['age']))
print("Медиана меньшее среднее значение: ", st.median_low(df['age']))
print("Медиана большее среднее значение: ", st.median_high(df['age']))

print("\nМода: ")
print('Значение моды: ', st.mode(df['age']))

print("\nМинимум: ", np.min(df['age']))
print("Максимум: ", np.max(df['age']))
```

```python
print("\nМатематическое ожидание: ")
print(np.nanmean(df['bmi']))

print("\nСредневзвешенное значение: ")
print(np.average(df['bmi']))

print("\nМедиана: ", st.median(df['bmi']))
print("Медиана меньшее среднее значение: ", st.median_low(df['bmi']))
print("Медиана большее среднее значение: ", st.median_high(df['bmi']))

print("\nМода: ")
print('Значение моды: ', st.mode(df['bmi']))

print("\nМинимум: ", np.min(df['bmi']))
print("Максимум: ", np.max(df['bmi']))
```

```python
# Метрики оценки вариативности данных

print("\nДисперсия: ")
print(st.pvariance(df['avg_glucose_level']))
print(st.pvariance(df['avg_glucose_level'], mu=st.mean(df['avg_glucose_level'])))

print("\nСтандартное отклонение: ")
print(st.pstdev(df['avg_glucose_level']))

print("\nКвантиль: ")
print(np.percentile(df['avg_glucose_level'], 5))
print(np.percentile(df['avg_glucose_level'], 95))
print(np.percentile(df['avg_glucose_level'], [25, 50, 75]))

print("\nРазмах: ")
duration_range = max(df['avg_glucose_level']) - min(df['avg_glucose_level'])
print(duration_range)
x = [i for i in range(0, len(df['avg_glucose_level']))]
y = df.loc[:, 'avg_glucose_level']
plt.xlabel('Номер человека')
plt.ylabel('Уровень глюкозы')
plt.title('Точечный график зависимости двух выборок x и y')
plt.grid(True)
plt.scatter(x, y, marker='.')
plt.show()
```

```python
print("\nДисперсия: ")
print(st.pvariance(df['age']))
print(st.pvariance(df['age'], mu=st.mean(df['age'])))

print("\nСтандартное отклонение: ")
print(st.pstdev(df['age']))

print("\nКвантиль: ")
print(np.percentile(df['age'], 5))
print(np.percentile(df['age'], 95))
print(np.percentile(df['age'], [25, 50, 75]))

print("\nРазмах: ")
duration_range = max(df['age']) - min(df['age'])
print(duration_range)
x = [i for i in range(0, len(df['age']))]
y = df.loc[:, 'age']
plt.xlabel('Номер человека')
plt.ylabel('Возраст')
plt.title('Точечный график зависимости двух выборок x и y')
plt.grid(True)
plt.scatter(x, y, marker='.')
plt.show()
```

```python
# Метрики оценки корреляции или взаимозависимости

print("\nКорреляция: ")
glucose = df['avg_glucose_level']
bmi = df['bmi']
plt.title('Точечный график зависимости двух выборок x и y')
plt.scatter(glucose, bmi, marker='.')
plt.grid(True)
plt.xlabel('avg_glucose_level')
plt.ylabel('bmi')
plt.show()

print("\nМатрица корреляций: ")
data = df.iloc[:, [0, 2, 3, 4, 8, 9, 11]]
sns.heatmap(data.corr(), annot=True, cmap='RdYlGn', linewidths=0.2)
fig = plt.gcf()
fig.set_size_inches(12, 12)
plt.show()

print("\nКорреляция Пирсона: ")
corr, p_value = pearsonr(df['avg_glucose_level'], df['age'])
print(corr, p_value)

print("\nКорреляция Спирмена: ")
corr, p_value = spearmanr(df['avg_glucose_level'], df['age'])
print(corr, p_value)

print("\nКорреляция Тау Кендалла: ")
corr, p_value = kendalltau(df['avg_glucose_level'], df['age'])
print(corr, p_value)

print("\nВыбросы: ")
print(df.isnull().sum())
glucose = df['avg_glucose_level']
bmi = df['bmi']
fig, ax = plt.subplots()
ax.scatter(glucose, bmi, c='blue', alpha=0.3, edgecolors='none')
ax.grid(True)
plt.xlabel('avg_glucose_level')
plt.ylabel('bmi')
plt.show()

glucose_new = glucose[bmi < 40]
bmi_new = bmi[bmi < 40]
fig, ax = plt.subplots()
ax.scatter(glucose_new, bmi_new, c='blue', alpha=0.3, edgecolors='none')
ax.grid(True)
plt.xlabel('glucose_new')
plt.ylabel('bmi_new')
plt.show()
```
