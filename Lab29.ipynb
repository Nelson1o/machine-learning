```python
import pandas as pd
import numpy as np
from apyori import apriori

# import warnings

# warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('E:\\Магистратура\\3\\Машинное обучение\\Market_Basket_Optimisation.csv')
df
```

```python
df = df.ffill(axis=1)
df
```

```python
transactions = []
for i in range(0, 7500): 
    transactions.append([str(df.values[i,j]) for j in range(0, 20)])
```

```python
result = list(apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 4, min_length = 2))
```

```python
association_results = list(result)
```

```python
rules=[]
for item in result:

    # первый индекс внутреннего списка
    # Содержит базовый предмет и добавляет предмет
    pair = item[0] 
    items = [x for x in pair]
    v1=str(items[0]) 
    v2= str(items[1])

    #второй индекс внутреннего списка
    v3= str(item[1])

    #третий индекс списка, расположенный на 0-м месте
    #третьего индекса внутреннего списка

    v4= str(item[2][0][2])
    v5=str(item[2][0][3])
    rows=(v1,v2,v3,v4,v5)
    rules.append(rows)
label=['title1','title2','support','confidence','lift']  
nm=pd.DataFrame.from_records(rules,columns=label)
nm
```
